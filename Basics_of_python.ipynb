```python
import random
import time
import matplotlib.pyplot as plt

# Q1
L = [11, 12, 13, 14]

# (i)
L.extend([50, 60])

# (ii)
L = [x for x in L if x not in [11, 13]]

# (iii)
L.sort()

# (iv)
L.sort(reverse=True)

# (v)
print("13 in L:", 13 in L)

# (vi)
print("Number of elements in L:", len(L))

# (vii)
print("Sum of all elements in L:", sum(L))

# (viii)
print("Sum of odd numbers in L:", sum(x for x in L if x % 2 != 0))

# (ix)
print("Sum of even numbers in L:", sum(x for x in L if x % 2 == 0))

# (x)
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

print("Sum of prime numbers in L:", sum(x for x in L if is_prime(x)))

# (xi)
L.clear()

# (xii)
del L

# Q2
D = {1: 5.6, 2: 7.8, 3: 6.6, 4: 8.7, 5: 7.7}

# (i)
D[8] = 8.8

# (ii)
D.pop(2, None)

# (iii)
print("6 in D:", 6 in D)

# (iv)
print("Number of elements in D:", len(D))

# (v)
print("Sum of values in D:", sum(D.values()))

# (vi)
D[3] = 7.1

# (vii)
D.clear()

# Q3
S1 = {10, 20, 30, 40, 50, 60}
S2 = {40, 50, 60, 70, 80, 90}

# (i)
S1.update({55, 66})

# (ii)
S1 -= {10, 30}

# (iii)
print("40 in S1:", 40 in S1)

# (iv)
print("Union of S1 and S2:", S1.union(S2))

# (v)
print("Intersection of S1 and S2:", S1.intersection(S2))

# (vi)
print("S1 - S2:", S1 - S2)

# Q4
# (i)
random_strings = [''.join(random.choices('abcdefghijklmnopqrstuvwxyz', k=random.randint(6, 8))) for _ in range(100)]
print("100 random strings:", random_strings)

# (ii)
prime_numbers = [x for x in range(600, 801) if is_prime(x)]
print("Prime numbers between 600 and 800:", prime_numbers)

# (iii)
divisible_by_7_and_9 = [x for x in range(100, 1001) if x % 7 == 0 and x % 9 == 0]
print("Numbers divisible by 7 and 9 between 100 and 1000:", divisible_by_7_and_9)

# Q5
list1 = random.sample(range(10, 31), 10)
list2 = random.sample(range(10, 31), 10)

print("Common numbers:", set(list1) & set(list2))
print("Unique numbers:", set(list1) ^ set(list2))
print("Minimum in both lists:", min(min(list1), min(list2)))
print("Maximum in both lists:", max(max(list1), max(list2)))
print("Sum of both lists:", sum(list1) + sum(list2))

# Q6
random_numbers = random.sample(range(100, 901), 100)
odd_numbers = [x for x in random_numbers if x % 2 != 0]
even_numbers = [x for x in random_numbers if x % 2 == 0]
prime_numbers = [x for x in random_numbers if is_prime(x)]

print("Odd numbers count:", len(odd_numbers))
print("Even numbers count:", len(even_numbers))
print("Prime numbers count:", len(prime_numbers))

# Q7
D = {1: "One", 2: "Two", 3: "Three", 4: "Four", 5: "Five"}
with open('dictionary.txt', 'w') as f:
    for key, value in D.items():
        f.write(f"{key}, {value}\n")

# Q8
L = ["One", "Two", "Three", "Four", "Five"]
with open('string_lengths.txt', 'w') as f:
    for item in L:
        f.write(f"{item}, {len(item)}\n")

# Q9
with open('random_strings.txt', 'w') as f:
    for _ in range(100):
        f.write(''.join(random.choices('abcdefghijklmnopqrstuvwxyz', k=random.randint(10, 15))) + '\n')

# Q10
with open('prime_numbers.txt', 'w') as f:
    for num in range(600, 801):
        if is_prime(num):
            f.write(str(num) + '\n')

# Q11
def time_function(func):
    start_time = time.time()
    func()
    end_time = time.time()
    return end_time - start_time

# Q12
def sort_list(n):
    return sorted(random.sample(range(1000000), n))

sizes = [5000, 10000, 15000, 20000, 25000]
times = []

for size in sizes:
    times.append(time_function(lambda: sort_list(size)))

plt.plot(sizes, times)
plt.xlabel('Number of elements')
plt.ylabel('Time taken (seconds)')
plt.title('Sorting performance')
plt.show()

# Q13
students = {
    'Alice': {'Math': 85, 'Science': 90, 'English': 92, 'History': 88, 'Art': 95},
    'Bob': {'Math': 78, 'Science': 85, 'English': 80, 'History': 92, 'Art': 87},
    'Charlie': {'Math': 92, 'Science': 88, 'English': 95, 'History': 90, 'Art': 85}
}

averages = {student: sum(marks.values()) / len(marks) for student, marks in students.items()}
max_student = max(averages, key=averages.get)
min_student = min(averages, key=averages.get)

print(f"Student with maximum average marks: {max_student}")
print(f"Student with minimum average marks: {min_student}")
```
